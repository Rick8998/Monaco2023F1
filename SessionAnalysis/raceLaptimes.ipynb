```python
import fastf1
import fastf1.plotting
import seaborn as sns
from matplotlib import pyplot as plt

# The misc_mpl_mods option enables minor grid lines which clutter the plot
fastf1.plotting.setup_mpl(misc_mpl_mods=False)
race = fastf1.get_session(2023, "Monaco", 'R')
race.load()
```

```python
driver= "VER"
driver_laps = race.laps.pick_driver(driver).reset_index()
```

```python
0 in driver_laps.index
```

```python
from matplotlib.lines import Line2D
import matplotlib.patches as patches
fig, ax = plt.subplots(figsize=(10, 6))
colors = fastf1.plotting.COMPOUND_COLORS
#del colors["UNKNOWN"]
ax.scatter(driver_laps["LapNumber"], driver_laps["LapTime"], c=driver_laps['Compound'].map(colors), marker="o")
handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor=v, label=k, markersize=8) for k, v in colors.items()]
ax.legend(title='color', handles=handles, bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(color="grey", linestyle=":")
ax.add_patch(patches.Rectangle((51, 0.00088), 27, 0.00065,  linewidth=1, facecolor='blue', alpha=0.2))
ax.text(69, 0.00148, "RAIN", bbox=dict(boxstyle='round', facecolor='blue', alpha=0.7), fontsize=15)
ax.set_title(f"{race.event.year} {race.event.EventName} - {race.name}\n{driver} lap time evolution on different compounds")
ax.set_xlabel("Lap", fontsize=15)
ax.set_ylabel("Laptime (minutes)", fontsize=15)
```
